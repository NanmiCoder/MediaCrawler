```python
from playwright.async_api import async_playwright

async def initialize_playwright():
    # 初始化 Playwright 实例和浏览器上下文
    playwright = await async_playwright().start()
    browser = await playwright.chromium.launch(headless=False)
    return playwright, browser

async def perform_actions(browser):
    # 在这里执行页面的具体操作
    page = await browser.new_page()
    await page.goto('https://example.com')
    # ... more actions

async def cleanup_playwright(playwright, browser):
    # 清理并关闭浏览器和 Playwright 实例
    await browser.close()
    await playwright.stop()

# 使用拆分后的函数执行
playwright, browser = await initialize_playwright()
await perform_actions(browser)
await cleanup_playwright(playwright, browser)
```
